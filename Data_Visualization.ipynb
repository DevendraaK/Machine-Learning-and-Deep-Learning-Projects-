DATA VISUALIZATION


```
import matplotlib.pyplot as plt
import numpy as np
```

```
x=np.arange(1,50,0.1)
y=np.sin(x)

plt.plot(x,y,'+',color='yellow',label='devendra')
plt.plot(x,np.cos(x),'>',color='blue',label='cosine')
plt.title("Sample graph")
plt.xlabel('x-axis')
plt.ylabel('y_axis')
plt.legend()
plt.show()
```

```
plt.plot(x,y)
plt.show()
```

```
x=[3,7,9,4,6]
y=[12,6,9,8,4]
y1=[3,5,8,4,7]
plt.scatter(x,y,marker='+',s=200,alpha=0.9,color='green',label='points')
plt.scatter(x,y1,marker='^',s=150,color='blue',label = 'triangle')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('Scatter-Plot')
plt.legend(loc='right')
plt.show()
```
